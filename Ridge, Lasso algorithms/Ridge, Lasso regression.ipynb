```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header = 1)
```

```python
df.head()
```

```python
df.info()
```

## Data cleaning

```python
## missing values 
#df.isnull().sum()
df[df.isnull().any(axis = 1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values 
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum() ## checking if there is still any null values
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns ## we can check the spaces in the column name
```

```python
## Fix spaces in columns name
df.columns = df.columns.str.strip()
df.columns
```

## Change the required columns as integer Datatype

```python
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int) 
```

```python
df.info()
```

## Changing the other columns to float datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i !='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## let save the cleaned dataset
df.to_csv('Algerian_forest_fires_Clean_Dataset.csv',index = False)
```
