## K Means Clustering Implementation

```python
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
import pandas as pd 
import numpy as np
%matplotlib inline
```

```python
X,y = make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
## Standardization - Feature scaling technique
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3,random_state = 42)
```

```python
X_scaled_data = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow method to select the K value
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init="k-means++")
```

```python
kmeans.fit_predict(X_scaled_data)
```

```python
y_pred = kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

```python
## Silhoutee scoring 
from sklearn.metrics import silhouette_score
sc_coefficient = []
for k in range (2,11):
    kmeans = KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_scaled_data)
    score = silhouette_score(X_scaled_data,kmeans.labels_)
    sc_coefficient.append(score)
```

```python
sc_coefficient
```

```python
## plot elbow curve
plt.plot(range(2,11),sc_coefficient)
plt.xticks(range(2,11))
plt.xlabel("No. of Clusters")
plt.ylabel("Silhoutee Coeffecient")
plt.show()
```
