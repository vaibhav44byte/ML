```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_index = pd.read_csv("economic_index.csv")
```

```python
df_index.head()
```

```python
## drop unneccessary columns 
df_index.drop(columns=["Unnamed: 0","year","month"],axis=1,inplace=True)
```

```python
df_index.head()
```

```python
## check null values
df_index.isnull().sum()
```

```python
## Visualization
import seaborn as sns 
sns.pairplot(df_index)
```

```python
df_index.corr() ## correlation
```

```python
## visualize data points more closely
plt.scatter(df_index['interest_rate'],df_index['unemployment_rate'],color='r')
plt.xlabel('Interest Rate')
plt.ylabel('Unemployment Rate')
```

```python
## independent and dependent features
X = df_index[['interest_rate','unemployment_rate']]
## we can write as X = df_index[:,:-1]
y = df_index['index_price']
```

```python
X.head()
```

```python
y
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
sns.regplot(
    x=df_index['interest_rate'],
    y=df_index['unemployment_rate']
)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
## cross validation
from sklearn.model_selection import cross_val_score
validation_score = cross_val_score(reg,X_train,y_train,scoring='neg_mean_squared_error'
                                  ,cv = 3)
```

```python
np.mean(validation_score)
```

```python
## prediction 
y_pred = reg.predict(X_test)
y_pred
```

```python
## performance Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print(score)
```
