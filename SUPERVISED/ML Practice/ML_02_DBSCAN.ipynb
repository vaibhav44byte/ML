## Practice for DBSCAN Clustering 

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X, y = make_circles(
    n_samples=500,
    factor=0.5,
    noise=0.05,
    random_state=42
)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## DBSCAN Algo
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)
```

```python
dbscan.fit(X_scaled)
```

```python
import numpy as np

labels = dbscan.labels_

print("Unique labels:", np.unique(labels))
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1], c= dbscan.labels_)
```
